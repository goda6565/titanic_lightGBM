```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
#データ読み込み
train = pd.read_csv("../dataset/train.csv")
train
```

```python
target = train[["Survived","Pclass","Fare"]]
target
```

```python
x_target = target[["Pclass","Fare"]]
t_target = target["Survived"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pclassに対するSurvivedのプロット
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=target)
plt.title('Survived count by Pclass')
plt.show()

# Fareに対するSurvivedのプロット
plt.figure(figsize=(6, 4))
sns.histplot(x='Fare', hue='Survived', data=target, kde=True)
plt.title('Survived distribution by Fare')
plt.show()
```

```python
plt.figure()
# モデルの訓練
clf = DecisionTreeClassifier(criterion="gini").fit(x_target, t_target)
plt.figure(figsize=(12, 12))
plot_tree(clf, filled=True)
plt.show()
```
