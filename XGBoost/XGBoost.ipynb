```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#データ読み込み
train = pd.read_csv("../dataset/train.csv")
test = pd.read_csv("../dataset/test.csv")
```

```python
y = train['Survived']
```

```python
X_merge = pd.concat([train.drop(["Survived"],axis = 1),test])
```

```python
X_merge = X_merge.drop(["Name","Ticket"],axis = 1)
```

```python
X_merge = pd.get_dummies(X_merge, drop_first=True)
```

```python
X_merge
```

```python
X = X_merge[:891]
```

```python
x_output = X_merge[891:]
```

```python
x_output
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_estimators=500,learning_rate=0.1,max_depth=6,early_stopping_rounds=10)
```

```python
eval_set = [(X_test, y_test)]
model.fit(X_train, y_train, eval_set=eval_set, verbose=True)
```

```python
from sklearn import metrics
y_pred = model.predict_proba(X_test)
metrics.log_loss(y_test, y_pred)
```

```python
y_pred = np.where(y_pred >= 0.5, 1, 0)
```

```python
y_pred = np.argmax(y_pred, axis=1)
```

```python
y_test = y_test.values
```

```python
# 精度
a = 0
for i in range(len(y_pred)):
    if y_pred[i] == y_test[i]:
        a += 1

acc = a / len(y_pred)
print("Accuracy: ", acc)
```
